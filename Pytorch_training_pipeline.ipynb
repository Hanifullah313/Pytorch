```python
import pandas as pd
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns =["id","Unnamed: 32"] , inplace= True)
```

```python
df.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], random_state=2)
```

```python
y_test.unique()
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_test
```

```python
tensor_X_train = torch.DoubleTensor(x_train)
tensor_X_test = torch.DoubleTensor(x_test)
tensor_y_train = torch.LongTensor(y_train)
tensor_y_test = torch.LongTensor(y_test)
```

```python
tensor_y_test
```

```python
# Defining the model
class My_model () :
  def __init__(self ,x):
    self.model_wieghts = torch.rand(tensor_X_train.shape[1],1 , requires_grad=True , dtype=torch.float64)
    self.model_bias = torch.rand(1 , requires_grad=True , dtype=torch.float64)

  def forward_pass(self , x):
    z=torch.matmul(x,self.model_wieghts)+self.model_bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

     # Calculate loss
    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
lr = 0.01
epochs = 25

model = My_model(tensor_X_train)
for epoch in range(epochs):
  y_pred = model.forward_pass(tensor_X_train)
  loss = model.loss_function(y_pred , tensor_y_train)
  loss.backward()
  with torch.no_grad():
    model.model_wieghts -= lr * model.model_wieghts.grad
    model.model_bias -= lr * model.model_bias.grad
    model.model_wieghts.grad.zero_()
    model.model_bias.grad.zero_()
  print(f"Epochs ,{epoch+1} , Loss ,{loss.item()}")
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward_pass(tensor_X_test)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == tensor_y_test).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
